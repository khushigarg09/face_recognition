```python

import cv2
import os
import uuid
```

```python


# Define the path where images will be saved
NEW_IMAGES_PATH = r"D:\code\AI PROJECTS\face_recognition_final\new_images"

# Ensure the directory exists
os.makedirs(NEW_IMAGES_PATH, exist_ok=True)

# Establish a connection to the external webcam
cap = cv2.VideoCapture(1)
while cap.isOpened(): 
    ret, frame = cap.read()
    if not ret:
        print("Failed to capture image")
        break

    # Cut down frame to 250x250px
    frame = frame[120:120+250, 200:200+250, :]

    if cv2.waitKey(1) & 0xFF == ord('a'):
        imgname = os.path.join(NEW_IMAGES_PATH, f'{uuid.uuid1()}.jpg')
        cv2.imwrite(imgname, frame)
        print(f"Image saved at: {imgname}")
    
    # Show image back to screen
    cv2.imshow('Image Collection', frame)
    
    # Breaking gracefully
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam
cap.release()
# Close the image show frame
cv2.destroyAllWindows()
```
